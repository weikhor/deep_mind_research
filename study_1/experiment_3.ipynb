```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
alpha = 0.015
total_steps = 250
```

```python
x_linspace = np.linspace(-6, 6, 1500)
y_linspace = np.linspace(-6, 6, 1500)
X, Y = np.meshgrid(x_linspace, y_linspace)

L = (X**2 - Y**2)

# 3D contour plot
fig = plt.figure(figsize=(8,7), constrained_layout=True)
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.contour3D(X, Y, L, 50)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.text2D(0.95, 0.5, "Loss", transform=ax.transAxes, 
          fontsize=12, fontweight='bold', rotation=-90, va='center')
plt.subplots_adjust(right=0.85)
plt.show()
```

```python
def loss_function(theta):
    x, y = theta
    return x**2 + y**2

def grad_function(theta):
    x, y = theta
    return np.array([2*x, 2*y])

def gd_theta_update(theta,alpha):
    new_update = alpha*grad_function(theta)
    new_theta = theta - new_update

    return new_theta

def sgd_theta_update(theta, alpha):
    new_noisy_update = alpha*grad_function(theta) + np.random.normal(0, 1, 2)
    new_theta = theta - new_noisy_update
    
    return new_theta
```
